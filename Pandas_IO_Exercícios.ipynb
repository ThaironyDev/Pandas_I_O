##Para saber mais: erro de encoding - Como resolver?

Quando tentamos abrir um arquivo que foi salvo com um tipo de formatação de caracteres diferente do padrão universal UTF-8, podemos ter um erro de encoding. Para resolver essa questão, podemos usar a biblioteca chardet para detectar o encoding de um arquivo. Podemos ver que o tipo de codificação do arquivo é 'ascii' que por sua vez é compatível com a 'UTF-8'.





```python
import chardet
import requests
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data_ponto_virgula.csv'
response = requests.get(url)
if response.status_code == 200:
  content = response.content
  encoding_info = chardet.detect(content)
  print(encoding_info)
else:
  print('Falha ao obter o aquivo da URL.')
```

##Desafio: normalizando o JSON obtido na API

>Para saber mais: obtendo arquivos JSON de APIs

---
Primeiramente acessamos os dados via requisição HTTP para a API da fruityvice.com. Depois, verificamos o status da requisição. Em seguida, instanciamos o conteúdo de interesse em uma variável. Usando a biblioteca json, carregamos esses dados que estão em formato JSON para um formato de dicionário Python. Por fim, usando o método pd.json_normalize, criamos um DataFrame normalizado a partir deste dicionário e apresentamos suas 5 primeiras linhas.

```python
import requests, json, pandas as pd
```

```python
url = 'https://fruityvice.com/api/fruit/all'
response = requests.get(url)
response.status_code
content = response.content
dados_dict = json.loads(content)
dados_frutas = pd.json_normalize(dados_dict)
dados_frutas.head()
```

##Desafio: lendo uma tabela de uma página web

> Neste caso após importar as libs de interesse realizamos a requisição http, depois de posse do conteúdo html instanciamos a variável dados_habitantes com a tabela de interesse usando o método pd.read_html()[], por fim eliminamos a coluna adicional correspondentes aos índices do arquivo e apresentamos as 5 primeiras linhas.



```python
import requests, pandas as pd
url = 'https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_popula%C3%A7%C3%A3o'
response = requests.get(url)
response.status_code
content = response.content
dados_habitantes = pd.read_html(content)[0]
dados_habitantes.drop('Unnamed: 0', axis=1, inplace=True)
dados_habitantes.head()
```

>Usando as ferramentas do sqlalchemy para criação e gestão do banco de dados.

```python
from sqlalchemy import create_engine, inspect
```

>Criando um banco de dados local.

---

>Criando o nosso motor de conexão com o banco de dados para isso vamos usar a função create_engine() passando uma string que informa qual SGBD vamos usar e onde será alocada.

---

'sqlite -> SGBD' eo ':///:mermory:' diz respeito a uma espaço na memória local.

```python
engine = create_engine('sqlite:///:memory')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dados = pd.read_csv(url)
dados.to_sql('clientes', engine, index=False)
```

>Verificando se a tabela foi criada com exito.

```python
inspetor = inspect(engine)
inspetor.get_table_names() # Exibe os nomes das tabelas que constam no banco de dados.
```

>Fazendo uma consulta ao banco de dados.

```python
query = 'SELECT * FROM clientes'
pd.read_sql(query, con=engine).head()
#ou
pd.read_sql_table('clientes', engine).head()
#Note que na primeira temos uma consulta SQL e na segunda temos um metodo onde informamos a tabela que queremos visualizar.
```

>Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.

```python
query = 'UPDATE clientes SET Rendimento_anual = 300000 WHERE ID_Cliente == 6840104'
with engine.connect() as conn:
  conn.execute(query)
query = 'SELECT *  FROM clientes WHERE ID_Cliente == 6840104'
pd.read_sql(query, con=engine).head()
```

>Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.

```python
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008809'
with engine.connect() as conn:
  conn.execute(query)
pd.read_sql_table('clientes', engine).head(6)
```
