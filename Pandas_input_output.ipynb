##Lendo dados no Google Planilhas

>Acessando a planilha que esta no Google planilhas usando um link de compatilhamento, em seguida, com este link podemos manipular essa planilha no colab, para isso √© necessario mudar o modo de compartilhamento do link no Google planilha.

>Para fazer esse procedimento precisaremos identificar parte do link que faz refer√™ncia ao nosso id 'https://docs.google.com/spreadsheets/d/1h6j5R5LlfVUDN-E4hSUFZBAeC5WKSU3332Uxcl_gwR0/edit?usp=sharing' o que estiver entre d/ conte√∫do /edit vai ser a parte que vamos precisar pois ela ser refere ao id da planilha.


```python
sheet_id = '1h6j5R5LlfVUDN-E4hSUFZBAeC5WKSU3332Uxcl_gwR0' # id da planilha
```

>Nesse momento usamos o fstring para compor o nossa URL, no da URL acrescentaremos um par√¢metro /gviz/tq que permite acessar a API do Google Planilhas, al√©m desse par√¢metro usaremos outro para que os dados sejam retornados no formato csv, que √© o seguinte ?tqx=out:csv , por fim temos o &sheet que faz com que todos os dados da planilha sejam retornados incluindo caba√ßalhos.

> Portanto a URL fica assim: url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'

```python
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

> Para facilitar todo esse processo e reutiliza√ß√£o do c√≥digo podemos criar uma fun√ß√£o.

```python
def planilha_do_google(url, sheet_name=''):
    """
    Esta fun√ß√£o foi projetada para receber 2 par√¢metros
    1¬∫ Link de compartilhamento da planilha do Google Planilhas que deve ter o status de acesso geral.
    2¬∫ Nome da p√°gina do arquivo Excel que estamos acessando. OBS: √© poss√≠vel informar a posi√ß√£o num√©rica da p√°gina iniciando em 0, caso n√£o queira informar o nome.

    Caso o segundo par√¢metro n√£o seja especificado, o retorno ser√° todo o arquivo Excel. Se o segundo par√¢metro for especificado, o retorno ser√° a p√°gina desejada.

    Se o link n√£o possuir um ID v√°lido, uma mensagem de erro ser√° exibida.
    """
    import re
    # Use uma express√£o regular para encontrar o que est√° entre "d/" e "/edit" e a parte antes de "d/"
    match = re.search(r'(.*?d/)([^/]+)/edit', url)

    if match:
        url = match.group(1)
        sheet_id = match.group(2) + '/gviz/tq?tqx=out:csv&sheet'
        url = url + sheet_id
        if sheet_name is not None:
            if sheet_name == 0:
                url = f'{url}=0'
            else:
                url = f'{url}={sheet_name}'
        return url
    else:
        print("ID n√£o encontrado na URL.")

url = 'https://docs.google.com/spreadsheets/d/1h6j5R5LlfVUDN-E4hSUFZBAeC5WKSU3332Uxcl_gwR0/edit?usp=sharing'
url = planilha_do_google(url, 'fontes')
pd.read_csv(url).head()
```

>Eliminando um registro do banco de dados a partir de uma condi√ß√£o. Nesse sentido com a classe do objeto engine que √© o connect usamos o m√©todo execute e executamos uma instru√ß√£o SQL ao final da execu√ß√£o a conex√£o com o banco de dados ser√° fechada.
