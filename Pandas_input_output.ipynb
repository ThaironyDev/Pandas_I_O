##Lendo dados no Google Planilhas

>Acessando a planilha que esta no Google planilhas usando um link de compatilhamento, em seguida, com este link podemos manipular essa planilha no colab, para isso é necessario mudar o modo de compartilhamento do link no Google planilha.

>Para fazer esse procedimento precisaremos identificar parte do link que faz referência ao nosso id 'https://docs.google.com/spreadsheets/d/1h6j5R5LlfVUDN-E4hSUFZBAeC5WKSU3332Uxcl_gwR0/edit?usp=sharing' o que estiver entre d/ conteúdo /edit vai ser a parte que vamos precisar pois ela ser refere ao id da planilha.


```python
sheet_id = '1h6j5R5LlfVUDN-E4hSUFZBAeC5WKSU3332Uxcl_gwR0' # id da planilha
```

>Nesse momento usamos o fstring para compor o nossa URL, no da URL acrescentaremos um parâmetro /gviz/tq que permite acessar a API do Google Planilhas, além desse parâmetro usaremos outro para que os dados sejam retornados no formato csv, que é o seguinte ?tqx=out:csv , por fim temos o &sheet que faz com que todos os dados da planilha sejam retornados incluindo cabaçalhos.

> Portanto a URL fica assim: url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'

```python
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

```python
dados_co2_sheets = pd.read_csv(url)
```

```python
dados_co2_sheets.head()
```

> Para facilitar todo esse processo e reutilização do código podemos criar uma função.

```python
def planilha_do_google(url, sheet_name=''):
    """
    Esta função foi projetada para receber 2 parâmetros
    1º Link de compartilhamento da planilha do Google Planilhas que deve ter o status de acesso geral.
    2º Nome da página do arquivo Excel que estamos acessando. OBS: é possível informar a posição numérica da página iniciando em 0, caso não queira informar o nome.

    Caso o segundo parâmetro não seja especificado, o retorno será todo o arquivo Excel. Se o segundo parâmetro for especificado, o retorno será a página desejada.

    Se o link não possuir um ID válido, uma mensagem de erro será exibida.
    """
    import re
    # Use uma expressão regular para encontrar o que está entre "d/" e "/edit" e a parte antes de "d/"
    match = re.search(r'(.*?d/)([^/]+)/edit', url)

    if match:
        url = match.group(1)
        sheet_id = match.group(2) + '/gviz/tq?tqx=out:csv&sheet'
        url = url + sheet_id
        if sheet_name is not None:
            if sheet_name == 0:
                url = f'{url}=0'
            else:
                url = f'{url}={sheet_name}'
        return url
    else:
        print("ID não encontrado na URL.")

url = 'https://docs.google.com/spreadsheets/d/1h6j5R5LlfVUDN-E4hSUFZBAeC5WKSU3332Uxcl_gwR0/edit?usp=sharing'
url = planilha_do_google(url, 'fontes')
pd.read_csv(url).head()
```

>Eliminando um registro do banco de dados a partir de uma condição. Nesse sentido com a classe do objeto engine que é o connect usamos o método execute e executamos uma instrução SQL ao final da execução a conexão com o banco de dados será fechada. Isso tudo num contexto with que garante o fechamento correto do banco de dados.

```python
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
with engine.connect() as conn:
  conn.execute(query) # esse trecho serve para garantir que a conexão com o banco de dados seja fechada corretamente após a execução da consulta.
```

>Confirmando que o registro foi apagado com sucesso.

```python
pd.read_sql_table('clientes', engine)
```
